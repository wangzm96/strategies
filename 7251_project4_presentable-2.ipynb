# Delivery 1 (My Bid/Ask vs. Market Bid/Ask)
- The orders are categorized into 4 groups since they are centered on different times (we use a threshold of 2 minutes)
- We adopt 4 market making strategies and therefore 4 sets of My Quote
  - Baseline: Baseline model is a simple market-following strategy which set bid & ask price equal to latest market bid & ask price.
  - Adjusted Baseline: Baseline_adjusted take open interest into consideration.  It skew our bid-ask spread to balance open interest in order to avoid market taking.If open interest is negative, which means we do not expect another buy order. Then we just increase our ask. If open interest is positive, which means we do not expect another sell order, then we just decrease our bid.
  - Aggressive Baseline: Baseline_adjusted_2 is a more aggresive version of baseline_adjusted. It also consider open interest. If open interest is negative, we increase our bid to encourage sell order instead of increasing our ask. If open interest is positive, we decrease our ask to encourage buy order
  - Exponential Moving Average: EMA consider market trend besides open interest. We use exponential moving average to detect market trends and consider trends when we determine our bid & ask.

```python
def group_order(order_times, seconds_):
    
    interval = dt.timedelta(seconds = seconds_) # orders come within certain seconds\
                                                  #are considered to be in the same order block
    idx_within_group = [0]
    idx_all_groups = []

    idx = 1
    while idx < len(order_times):
        if order_times[idx] - order_times[idx-1] < interval:
            idx_within_group.append(idx)
        else:
            idx_all_groups.append(idx_within_group)
            idx_within_group = [idx]
    
        idx += 1
    
    idx_all_groups.append(idx_within_group)
    
    return idx_all_groups
```

```python
def insert_order(df_orders, idx, succession_time): # succession time in unit of millisecond    
    signal = df_orders['Side'][idx]
    current_time = df_orders.index[idx]
    time_delta = dt.timedelta(milliseconds=succession_time)
    target_time = current_time + time_delta
    
    above = df_orders[:idx+1]
    below = df_orders[idx+1:]
    
    inserted = pd.DataFrame([(target_time, signal, 1)])
    inserted.set_index(keys = [0], inplace=True)
    inserted.index.name = 'Time'
    inserted.columns = ['Side','Follow-up']
    df_orders = pd.concat([above, inserted, below], axis=0)
    
    return df_orders  
```

```python
def order_process(info_order, info_mkt, df_my_quotes, q, p, X):
    '''
    info_order is a slice of df_order; eg: info_orders = df_orders.iloc[0]
    info_order is a slice of latest df_mkt; eg: info_orders = df_mkt.iloc[1]
    df_my_quotes is a dataframe of my quotes; index is a series of time; columns are my bids and asks for each time point
    p is the probalibility of an order to be filled by me if my price is the same as market price
    X is the reaction time, in unit of milliseconds
    
    Return three integers, the first indicating to fill/not fill the order, the second indicating if follow-up order 
    will be possible, three being the index of my quote that was actually used
    '''
    signal = info_order[0]
    follow_up = info_order[1]
    order_time = info_order.name
    
    mkt_bid = info_mkt[0]
    mkt_ask = info_mkt[1]
    
    reaction_time = dt.timedelta(milliseconds = X)
    effective_quote_time = order_time - reaction_time
    idx_effective_quote = df_my_quotes.index.get_loc(effective_quote_time, method = 'ffill')
    effective_quote = df_my_quotes.iloc[idx_effective_quote]
    my_bid = effective_quote[0]
    my_ask = effective_quote[1]
    
    if follow_up == 1:
        if signal == 'B':
            return -1, 0, idx_effective_quote
        if signal == 'S':
            return 1, 0, idx_effective_quote
        
    elif follow_up == 0:
        if signal == 'B':    #market taker's buying
            if mkt_ask < my_ask:
                return 0, 0, idx_effective_quote
            elif mkt_ask > my_ask:
                return -1, np.random.binomial(1, q), idx_effective_quote
            elif mkt_ask == my_ask:
                return -np.random.binomial(1, p), 0, idx_effective_quote
        elif signal == 'S':
            if mkt_bid > my_bid:
                return 0, 0, idx_effective_quote
            elif mkt_bid < my_bid:
                return 1,  np.random.binomial(1, q), idx_effective_quote
            elif mkt_bid == my_bid:
                return np.random.binomial(1, p), 0, idx_effective_quote
    else:
        print("Error in order_process()")
        sys.exit(1)
```

```python
def check_bid_ask(my_bid,my_ask,open_interest):
    if my_bid <= my_ask : return my_bid, my_ask
    if my_bid > my_ask:
        if open_interest > 0 : return my_ask,my_ask
        elif open_interest <= 0 : return my_bid,my_bid
        
```

```python
def baseline(df_mkt_available,open_interest):
    my_bid = df_mkt_available.iloc[-1][0]
    my_ask = df_mkt_available.iloc[-1][1]
    my_quote_time = df_mkt_available.iloc[-1].name
    
    return my_bid, my_ask, my_quote_time
```

```python
def baseline_adjusted(df_mkt_available,open_interest):
    # skew 
    my_bid = df_mkt_available.iloc[-1][0]-abs(max(open_interest,0))*0.00001
    my_ask = df_mkt_available.iloc[-1][1]+abs(min(open_interest,0))*0.00001
    my_quote_time = df_mkt_available.iloc[-1].name
    return my_bid, my_ask, my_quote_time
```

```python
def baseline_adjusted_2(df_mkt_available,open_interest):
    # more aggresive strategy, may fail if there are continues buy/sell orders
    my_bid = df_mkt_available.iloc[-1][0]+abs(min(open_interest,0))*0.00001
    my_ask = df_mkt_available.iloc[-1][1]-abs(max(open_interest,0))*0.00001
    my_quote_time = df_mkt_available.iloc[-1].name
    return my_bid, my_ask, my_quote_time
```

```python
def trend_determine(series_):
    '''
    series_ is the difference of short-term series and long-term series (short - long)
    the return is the expectation of price rising(return 1) or falling(return -1)
    '''
    series_ = series_[series_!=0]
    length = len(series_)
    i = -2
    current = 0
    signal = 0
    while abs(i) <= length:
        if series_[i]>0 and series_[i+1]<0: # expect price to fall
            signal = -1
            break
        elif series_[i]<0 and series_[i+1]>0:
            signal = 1
            break
        i -= 1
        
    return signal
```

```python
def ema(df_mkt_available, open_interest):
    current_mkt_bid = df_mkt_available.iloc[-1][0]
    current_mkt_ask = df_mkt_available.iloc[-1][1]
    current_mkt_qt_time = df_mkt_available.iloc[-1].name
    
    if len(df_mkt_available) <= 20:
        my_bid, my_ask, my_quote_time = baseline_adjusted(df_mkt_available,open_interest)
        return my_bid, my_ask, my_quote_time
    else:
        stream_bid = df_mkt_available['Bid']
        stream_ask = df_mkt_available['Ask']
        stream_mid = (stream_bid + stream_ask)/2
        
        ema_short = stream_mid.ewm(span=5).mean()
        ema_long = stream_mid.ewm(span=20).mean()
        diff = ema_short - ema_long
        signal = trend_determine(diff)

        my_bid = current_mkt_bid + 2*signal*0.00001 + abs(min(open_interest,0))*0.00001
        my_ask = current_mkt_ask - 2*signal*0.00001 - abs(max(open_interest,0))*0.00001
        my_quote_time = current_mkt_qt_time

        return my_bid, my_ask, my_quote_time
```

```python
def open_interest_management(open_interest,open_interest_threshold,result_orders,mkt_quote_time,
                             time_this_order,mkt_bid, mkt_ask,my_bid,my_ask,history_bid,history_ask):
    if open_interest > open_interest_threshold:
        # market taker 
        open_interest -= 1
        result_orders.append((mkt_quote_time,time_this_order,mkt_bid,mkt_ask,my_bid,my_ask,"S_T",1,0,open_interest))
        history_bid.append(mkt_bid)
    if open_interest < -open_interest_threshold:
        open_interest += 1
        result_orders.append((mkt_quote_time,time_this_order,mkt_bid,mkt_ask,my_bid,my_ask,"B_T",1,0,open_interest))
        history_ask.append(mkt_ask)
    return open_interest,history_bid,history_ask

def dynamic_pnl_management(open_interest,result_orders,mkt_quote_time,mkt_bid, mkt_ask,my_bid,my_ask
                        ,history_bid,history_ask):
   
    if open_interest < 0:
        open_interest += 1
        result_orders.append((mkt_quote_time,mkt_quote_time,mkt_bid,mkt_ask,my_bid,my_ask,"B_T",1,0,open_interest))
        history_ask.append(mkt_ask)
    elif open_interest > 0:
        open_interest -= 1
        result_orders.append((mkt_quote_time,mkt_quote_time,mkt_bid,mkt_ask,my_bid,my_ask,"S_T",1,0,open_interest))
        history_bid.append(mkt_bid)
    return open_interest,history_bid,history_ask
```

```python
def CalRealizedPnL(df):
    '''
    input: grouped executed order
    
    output: dataframe with realized PnL
    '''
    df_origin = df.copy()
    df['trade'] = df['Side'].apply(lambda x: -1 if (x == "B" or x == "S_T") else 1)
    df["trade"] = df["trade"]*df["Status"]
    df.loc[df.trade == -1, "executed_sell"] = (-df["My_Ask"])*(df.trade)
    df.loc[df.trade == 1, "executed_buy"] = (-df["My_Bid"])*(df.trade)
    df.loc[df.Side == "B_T", "executed_buy"] = (-df["Mkt_Ask"])*(df.trade)
    df.loc[df.Side == "S_T", "executed_sell"] = (-df["Mkt_Bid"])*(df.trade)
    df["executed_sell"] = df["executed_sell"].fillna(0)
    df["executed_buy"] = df["executed_buy"].fillna(0)
    df["executed"] = df["executed_sell"]+df["executed_buy"]
    
    RealizedPnL = []
    RealizedTime = []
    buy = []
    buyidx = []
    sell = []
    sellidx = []
    for i in df.index:
        if df.executed[i] > 0:
            sell.append(df.executed[i])
            sellidx.append(i)
        elif df.executed[i] < 0:
            buy.append(df.executed[i])
            buyidx.append(i)
    
    length = np.minimum(len(sell), len(buy))
    for i in range(length):
        RealizedPnL.append(sell[i]+buy[i])
        RealizedTime.append(np.maximum(sellidx[i], buyidx[i]))
    df_RealizedPnL = pd.DataFrame(RealizedPnL, index = RealizedTime, columns = ["RealizedPnL"])
    df1 = pd.concat([df_origin,df_RealizedPnL], axis = 1)
    df1["RealizedPnL"] = df1.RealizedPnL.fillna(0)
    #df1["RealizedPnL"].cumsum().plot(grid=True)
    df1.loc[(df1.Side == "B_T"), "HedgeCost"] = -0.5*(df1["Mkt_Ask"] - df1["Mkt_Bid"])*1000000 #in dollars
    df1.loc[(df1.Side == "S_T"), "HedgeCost"] = -0.5*(df1["Mkt_Ask"] - df1["Mkt_Bid"])*1000000
    df1 = df1[['Mkt_Quote_Time','Order_Time','Mkt_Bid','Mkt_Ask',\
                            'My_Bid','My_Ask', 'Side', 'Status', 'Follow-up','Open_Interest','RealizedPnL','HedgeCost']]
    df1["RealizedPnL"] = df1["RealizedPnL"]*1000000
    return df1

def plot_pnl(result_allgroups,strategy):
    plt.subplots(2,2,figsize=(12,8))
    for i in range(len(result_allgroups)):
        tmp = CalRealizedPnL(result_allgroups[i])
        plt.subplot(2,2,i+1)
        tmp["RealizedPnL"].cumsum().plot(grid=True)
        plt.title("RealizedPnL for {}".format(strategy))
        
```

```python
def plot_tick_offer(strategy):
    for j in range(len(globals()["my_quotes_{}".format(strategy.__name__)])):
        globals()["my_quotes_{}".format(strategy.__name__)][j][["MyBid","MktBid"]].plot(figsize=(10,5),grid=True,title="Order Group {a}: MyBid & MktBid for {b}".format(a = j+1, b = strategy.__name__))
        globals()["my_quotes_{}".format(strategy.__name__)][j][["MyAsk","MktAsk"]].plot(figsize=(10,5),grid=True,title="Order Group {a}: MyAsk & MktAsk for {b}".format(a = j+1, b = strategy.__name__))

def plot_dynamic_pnl(strategy):
    for j in range(len(globals()["dynamic_pnl_{}".format(strategy.__name__)])):
        globals()["dynamic_pnl_{}".format(strategy.__name__)][j].plot(figsize=(10,5),grid=True)
```

```python
def cal_dynamic_pnl(history_bid,history_ask,mkt_bid,mkt_ask):
    paired = min(len(history_bid),len(history_ask))
    realized_pnl = sum(history_ask[:paired])-sum(history_bid[:paired])
    if len(history_bid)>len(history_ask):
        dynamic_pnl = mkt_ask*(len(history_bid)-paired) - sum(history_bid[paired:])
    if len(history_ask)>len(history_bid):
        dynamic_pnl = sum(history_ask[paired:]) - mkt_bid*(len(history_ask)-paired)
    if len(history_ask)==len(history_bid):dynamic_pnl=0
    return [(realized_pnl+dynamic_pnl)*1000000,dynamic_pnl*1000000]
```

```python
def stats(strategylist):
    returnlist = []
    for j in range(len(globals()["result_allgroups_{}".format(strategylist[0].__name__)])):
        MDD = []
        MeanPnL = []
        MedianPnL = []
        PnLVolatility = []
        HedgeCost = []
        ExecutedOrders = []
        TradeAsMT = []
        TotalOrders = []
        idx = []
        for strategy in strategylist:
            df = CalRealizedPnL(globals()["result_allgroups_{}".format(strategy.__name__)][j])
            MDD.append((df.RealizedPnL.cumsum() - df.RealizedPnL.cumsum().cummax()).min())
            MeanPnL.append(df.RealizedPnL.mean())
            MedianPnL.append(df.RealizedPnL.median())
            PnLVolatility.append(df.RealizedPnL.std())
            HedgeCost.append(df.HedgeCost.sum())
            ExecutedOrders.append(df.Status.sum())
            TradeAsMT.append(df["HedgeCost"].loc[df["HedgeCost"]<0].count())
            TotalOrders.append(df.Side.count())
            idx.append(strategy.__name__)
        returnlist.append(pd.DataFrame({"MDD($)":MDD, "MeanPnL($)":MeanPnL, "MedianPnL($)":MedianPnL, "PnLVolatility":PnLVolatility, "HedgeCost($)":HedgeCost, "TotalOrders":TotalOrders, "ExecutedOrders": ExecutedOrders, "TradeAsMT": TradeAsMT}, index = idx).round(2))
        
    return returnlist
```

```python
def market_making(strategy):
    count_start = dt.datetime.now()
    p = 0.5
    q = 0.25
    Y = 5
    X = 10
    loss_limit = -2500
    open_interest = 0
    open_interest_threshold = 3
    order_times = df_orders.index
    groups = group_order(order_times, 120)
    
    interval = dt.timedelta(seconds = 180) # look back and forward for 180s, respectively

    result_allgroups = []
    my_quotes = []
    dynamic_pnl = []
    
    
    for i in range(len(groups)):
        history_bid = []
        history_ask = []
        idx_first_order = groups[i][0]
        time_first_order = df_orders.index[idx_first_order]
        time_start = time_first_order - interval
    
        idx_last_order = groups[i][-1]
        time_last_order = df_orders.index[idx_last_order]
        time_end = time_last_order + interval

        idx_mkt_first = df_mkt.index.get_loc(time_start, method = 'nearest')
        idx_mkt_last = df_mkt.index.get_loc(time_end, method = 'nearest')

        df_orders_thisgroup = df_orders[idx_first_order : idx_last_order+1]
        df_mkt_thisgroup = df_mkt[idx_mkt_first : idx_mkt_last]
        
        k = 0
        info_this_order = df_orders_thisgroup.iloc[k]
        time_this_order = info_this_order.name
        df_my_quotes = pd.DataFrame(index = [], columns = ['MyBid','MyAsk','MktBid','MktAsk'])
        df_dynamic_pnl = pd.DataFrame(index = [],columns = ["Dynamic_PnL"])
        j = 1
        result_orders = []
        
        while j < len(df_mkt_thisgroup):
            df_mkt_available = df_mkt_thisgroup[:j]
            mkt_bid,mkt_ask = df_mkt_available.iloc[-1][0],df_mkt_available.iloc[-1][1]
            mkt_quote_time = df_mkt_available.iloc[-1].name
            my_bid, my_ask, my_quote_time = strategy(df_mkt_available,open_interest)
            my_bid, my_ask = check_bid_ask(my_bid,my_ask,open_interest)
            df_my_quotes.loc[my_quote_time] = [my_bid, my_ask,mkt_bid,mkt_ask]
            dynamic_pnl_this_time = cal_dynamic_pnl(history_bid,history_ask,mkt_bid,mkt_ask)
            df_dynamic_pnl.loc[my_quote_time] = [dynamic_pnl_this_time[0]]
            if dynamic_pnl_this_time[1]<loss_limit: 
                open_interest,history_bid,history_ask = dynamic_pnl_management(open_interest,result_orders,
                                                            mkt_quote_time,mkt_bid, mkt_ask,my_bid,my_ask,history_bid,history_ask)
            while df_mkt_thisgroup.index[j] <= time_this_order and df_mkt_thisgroup.index[j+1] > time_this_order: # an order arrives 
                info_mkt = df_mkt_available.iloc[-1]
                order_status, follow_up_possi, idx_my_quote = order_process(info_this_order, info_mkt, df_my_quotes,q, p, X)
                my_quote__ = df_my_quotes.iloc[idx_my_quote]
                my_bid__ = my_quote__[0]
                my_ask__ = my_quote__[1]
                if follow_up_possi == 1:
                    df_orders_thisgroup = insert_order(df_orders_thisgroup, k, Y)
                
                if order_status == -1 :history_ask.append(my_ask)
                if order_status == 1:history_bid.append(my_bid)
                    
                open_interest += order_status
                mkt_quote_time = info_mkt.name
                mkt_bid = info_mkt[0]
                mkt_ask = info_mkt[1]
                side = info_this_order[0]
                follow_up = info_this_order[1]
                result_orders.append((mkt_quote_time, time_this_order, mkt_bid, mkt_ask, my_bid__, my_ask__,\
                side, abs(order_status), follow_up,open_interest))

                open_interest,history_bid,history_ask = open_interest_management(open_interest,open_interest_threshold,result_orders,
                                                         mkt_quote_time,time_this_order,mkt_bid, mkt_ask,my_bid,my_ask
                                                        ,history_bid,history_ask)

                k += 1
                if k >= len(df_orders_thisgroup):
                    break
                info_this_order = df_orders_thisgroup.iloc[k]
                time_this_order = info_this_order.name


            j += 1

        df_result_orders = pd.DataFrame(result_orders)
        df_result_orders.columns = ['Mkt_Quote_Time','Order_Time','Mkt_Bid','Mkt_Ask',\
                                'My_Bid','My_Ask', 'Side', 'Status', 'Follow-up','Open_Interest']

        result_allgroups.append(df_result_orders)
        my_quotes.append(df_my_quotes)
        dynamic_pnl.append(df_dynamic_pnl)
        
    print("Time used for strategy: {}=".format(strategy.__name__), dt.datetime.now() - count_start)
    return result_allgroups,my_quotes,dynamic_pnl

```
